```python
artist, track_ids, tracks, features = [], [], [], []
i = 0
while i < len(track_list1):
    artist.append(track_list1[i]['track']['artists'])
    track_ids.append(track_list1[i]['track']['id'])
    tracks.append(track_list1[i]['track']['name'])
    features.append(spotify_obj.audio_features(track_ids[i])[0])
    i += 1
```

```python
def generalgenre(genrelist):
    if 'edm' in genrelist or 'pop edm' in genrelist:
        return 'edm'
    elif 'hip hop' in genrelist or 'rap' in genrelist:
        return 'hip hop'
    elif  ('pop' in genrelist or 'k-pop' in genrelist) and ('dance pop' in genrelist or 'pop rap' in genrelist) or ('dance pop' in genrelist and 'pop rap' in genrelist):
        return 'pop'
    elif 'r&b' in genrelist:
        return 'r&b'
    elif 'modern rock' in genrelist:
        return 'rock'
    for genre in genrelist:
        string = genre.split() 
        if string[-1] in ['brostep','house','disco','dance','electropop'] or string[0] == 'dance':
            return 'edm'
        elif string[-1] == 'hop' or genre.split()[-1] == 'rap' or genre.split()[0] == 'trap':
            return 'hip hop'
        elif string[-1] == 'rock':
            return 'rock'
        elif string[-1] == 'pop' or string[0] == 'pop':
            return 'pop'
    return 'pop'
        
def leadingartist(artistlist):
    lead = artistlist[0]
    for artist in artistlist:
        if spotify_obj.artist(artist['id'])['genres'] != []:
            return artist['name'], artist['id']
    return lead['name'], lead['id']
```

```python
df = pd.DataFrame(features, tracks)
df['artist'] = artist
df['artist'] = df['artist'].apply(leadingartist)
df['artist_id'] = df['artist'].apply(lambda x: x[1])
df['artist'] = df['artist'].apply(lambda x: x[0])
df.drop(['analysis_url','track_href', 'type','uri', 'time_signature'],axis = 1,inplace = True)
df.head()
```

```python
genre = []
for artist in df['artist_id']:
    genre.append(spotify_obj.artist(artist)['genres'])
df['genre'] = genre
df['genre'] = df['genre'].apply(generalgenre)
df.head()
```

```python
def fix_column(data, column, dictionary):
    data_copy = data.copy()
    data_copy[column] = data_copy[column].map(dictionary)
    return data_copy
    
fixed_df = fix_column(df, 'key', {
        0: 'C',
        1: 'C#/Db',
        2: 'D',
        3: 'D#/Eb',
        4: 'E',
        5: 'F',
        6: 'F#/Gb',
        7: 'G',
        8: 'G#/Ab',
        9: 'A',
        10: 'A#/Bb',
        11: 'B'
    })
fixed_df.head()
```

```python

def ohe_column(data, column):
    categories = pd.get_dummies(data[column])
    return pd.concat([data, categories], axis = 1)

ohe_df = ohe_column(fixed_df, 'genre')
ohe_df.head()
```

## Principal Component Analysis

```python
import matplotlib.pyplot as plt

X = ohe_df.drop(['id','artist', 'good', 'genre', 'key', 'artist_id'], axis = 1)
y = ohe_df['good']
df_means = np.mean(X, axis = 0)
normalization = (X - df_means)/np.std(X)
normalization.head()
```

```python
u, s, vt = np.linalg.svd(normalization, full_matrices = False)

X_new = pd.DataFrame(u * s)
X_new['name'] = normalization.index.values
X_new.set_index('name', inplace = True)
X_new.head()
```

```python
plt.figure(figsize=(10,7))

total_variance = np.sum(np.square(s))
perc = np.square(s) / total_variance

plt.plot(np.arange(1, 20), np.cumsum(perc))

plt.title("Number of Principal Components vs. Cumulative Variance")
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Variance')
plt.axvline(15)
plt.axhline(0.95)
plt.show()
```
