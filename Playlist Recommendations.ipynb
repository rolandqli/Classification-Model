```python
client_id = '7d731118e245470b827660ce69981b70'
client_secret = '034b90fa8e1c460f9db8f64c8c525a1d'
user = '12128951079'
scope = 'user-library-read playlist-modify-public playlist-read-private user-top-read'
redirect_uri = 'https://developer.spotify.com/dashboard/applications/7d731118e245470b827660ce69981b70'
```

```python
# Obtaining token for authorization
token = util.prompt_for_user_token(user, scope, client_id, client_secret, redirect_uri)

if token:
    spotify_obj = sp.Spotify(auth=token)
else:
    print("Can't get token for", username)
```

```python
source_playlist = spotify_obj.user_playlist(user, '3r7kDwRbuTBBdKKkSTAENl')
source_playlist
```

```python
track_list = source_playlist['tracks']['items']
track_list
```

```python
artist, track_ids, tracks, features = [], [], [], []
i = 0
while i < len(track_list):
    artist.append(track_list[i]['track']['artists'][0]['name'])
    artist_ids.append(track_list[i]['track']['artists'][0]['id'])
    track_ids.append(track_list[i]['track']['id'])
    tracks.append(track_list[i]['track']['name'])
    features.append(spotify_obj.audio_features(track_ids[i])[0])
    i += 1
```

```python
df = pd.DataFrame(features, tracks)
df['artist'] = artist
df.drop(['analysis_url','track_href', 'type','uri', 'time_signature'],axis = 1,inplace = True)
df.head()
```
